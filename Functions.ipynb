Write a function count_vowels(s) that counts and returns the number of vowels in string.

Yahan pe vowels = "aeiou" string ka use kiya gaya hai — tuple (('a','e','i','o','u')) ya list (['a','e','i','o','u']) ke bajaye.

1. String bhi iterable object hai — matlab iske har character ko ek-ek karke loop me access kiya ja sakta hai.

"aeiou" ek iterable object hai.

Hum isme loop chala sakte hain jaise list ya tuple me chalaate hain.

| Type       | Example                 | Use Case                                       |
| ---------- | ----------------------- | ---------------------------------------------- |
| **String** | `"aeiou"`               | Simple, readable, perfect for small fixed data |
| **List**   | `['a','e','i','o','u']` | Jab aapko elements modify karne hain           |
| **Tuple**  | `('a','e','i','o','u')` | Jab fixed data chahiye (immutable)             |
| **Set**    | `{'a','e','i','o','u'}` | Jab fast membership check chahiye (large data) |


```python
def count_vowels(s):
    vowels = "aeiou"
    count = 0
    for char in s.lower():
        if char in vowels:
            count += 1
    return count

input_string = input("Enter a string: ")
print("Number of vowels:", count_vowels(input_string))
```

```python
def factorial_iterative(n):
  if n < 0 :
    return "Invalid Number"
  result = 1
  for i in range(1, n+1):
    result *= i
  return result
print(factorial_iterative(5))
```

```python
def factorial(n):
  if n == 0 or n==1:
    return 1

  else:
    return n *factorial(n-1)
factorial(5)
```

```python
def fibonacci (n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fibonacci(n-1) + fibonacci(n-2)
print (fibonacci(7))
```

```python
def level_tracker(level):
    score = 0  # Enclosed variable for tracking score

    def update_score(points):
        nonlocal score
        score += points
        print(f"Level {level}: Score = {score}")

    return update_score

level1 = level_tracker(1)
level2 = level_tracker(2)

level1(100)  # Output: Level 1: Score = 100
level1(50)   # Output: Level 1: Score = 150
level2(200)  # Output: Level 2: Score = 200
```
