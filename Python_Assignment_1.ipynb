5. Explain the concept of typecasting in python with examples.
1. Implicit Typecasting (Automatic Conversion)

üëâ Python automatically converts one data type into another without user intervention.

Example:
a = 5       # int
b = 2.5     # float

result = a + b
print(result)
print(type(result))

Output:
7.5
<class 'float'>


üü¢ Explanation:
Python automatically converted the integer a into a float so that both operands are of the same type (float).
This is called implicit typecasting.

2. Explicit Typecasting (Manual Conversion)

üëâ You manually convert one data type into another using built-in functions like:

int()

float()

str()

bool()

list(), tuple(), etc.

Example 1: Converting float to int
x = 10.9
y = int(x)
print(y)
print(type(y))


Output:

10
<class 'int'>


üü† Decimal part is removed, not rounded.

Example 2: String to Integer
s = "123"
num = int(s)
print(num + 5)
print(type(num))


Output:

128
<class 'int'>


üü¢ String "123" is converted into integer 123.

Example 3: Integer to String
n = 45
s = str(n)
print("The number is " + s)


Output:

The number is 45

6. How do conditional statements work in python ? Illustrates with examples

Conditional statements are used to make decisions in your code ‚Äî they allow your program to execute different blocks of code based on certain conditions.

In simple words:

‚ÄúIf a condition is true, do something; otherwise, do something else.‚Äù

üîπ Main Conditional Statements in Python

if statement

if-else statement

if-elif-else ladder

Nested if statement

1Ô∏è‚É£ if Statement

üëâ Used to check a condition. If the condition is True, the block inside if executes.

age = 18

if age >= 18:
    print("You are eligible to vote.")


Output:

You are eligible to vote.


üü¢ Since age >= 18 is True, the statement runs.

2Ô∏è‚É£ if-else Statement

üëâ Runs one block if condition is True, otherwise another block.

age = 16

if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")


Output:

You are not eligible to vote.


üü† Since the condition is False, the else block runs.

3Ô∏è‚É£ if-elif-else Ladder

üëâ Used when you have multiple conditions to check.

marks = 75

if marks >= 90:
    print("Grade: A+")
elif marks >= 75:
    print("Grade: A")
elif marks >= 60:
    print("Grade: B")
else:
    print("Grade: C")


Output:

Grade: A


üü¢ Python checks each condition from top to bottom and executes the first one that‚Äôs True.

4Ô∏è‚É£ Nested if Statement

üëâ An if statement inside another if statement.

num = 10

if num > 0:
    if num % 2 == 0:
        print("Positive and Even number")
    else:
        print("Positive but Odd number")
else:
    print("Negative number")


Output:

Positive and Even number


üü¢ Both conditions are checked ‚Äî first whether the number is positive, then whether it‚Äôs even.

7. Describe the Different types of loops in python and their use cases with examples.

Types of Loops in Python

Loops are used in Python to repeat a block of code multiple times until a condition is met or a sequence ends.

There are two main types of loops in Python:

for loop

while loop

Additionally, Python supports loop control statements like break, continue, and pass.

üü¢ 1Ô∏è‚É£ for Loop

üëâ Used to iterate over a sequence (like a list, tuple, string, or range).

Syntax:

    # code block

‚úÖ Example 1: Print numbers from 1 to 5

    print(i)

2
3

üü¢ range(1, 6) generates numbers from 1 to 5.

‚úÖ Example 2: Iterate over a list

    print(fruit)

banana
mango


üìò Use case: When you want to loop through a collection of items (lists, strings, etc.).

üîµ 2Ô∏è‚É£ while Loop
